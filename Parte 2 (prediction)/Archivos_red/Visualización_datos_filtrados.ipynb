```python
import os
import matplotlib.pyplot as plt
import pandas as pd

# 1. Configuración de parámetros
FS = 60  # Frecuencia de muestreo (Hz)
WINDOW_DURATION = 2  # Duración de la ventana en segundos
SAMPLES_PER_WINDOW = FS * WINDOW_DURATION  # Muestras por ventana

# Ruta base proporcionada para guardar las figuras
OUTPUT_DIR = r"C:\Users\adria\OneDrive\Documentos\DATOS FILTRADOS\Recortes\20260423_subject_01_cond_ASS-PRE-EXT-RIGHT_run03_ecr_fcr_final_recortado"
BASE_NAME = "20260423_subject_01_cond_ASS-PRE-EXT-RIGHT_run03_ecr_fcr_final_recortado"

# Crear la carpeta de destino si no existe para evitar errores
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)

# 2. Cargar los datos
df = pd.read_csv(
    "C:/Users/adria/Downloads/Python-IMU/Adriana-Neuroengineering_lab/Archivos_red/Recortes/20260423_subject_01_cond_ASS-PRE-EXT-RIGHT_run03_ecr_fcr_final_recortado.csv"
)

# Creamos una columna de tiempo en segundos para el eje X
df["Time"] = df.index / FS

# 3. Iterar y graficar por ventanas
total_samples = len(df)
print(f"Total de muestras: {total_samples}")

for start_idx in range(0, total_samples, SAMPLES_PER_WINDOW):
    window_num = (start_idx // SAMPLES_PER_WINDOW) + 1
    end_idx = start_idx + SAMPLES_PER_WINDOW

    # Extraer el segmento actual de datos
    window_data = df.iloc[start_idx:end_idx]

    # Preparar el zoom del ángulo con un rango Y más estrecho
    angle_min = window_data["wrist_hand_r3_pos"].min()
    angle_max = window_data["wrist_hand_r3_pos"].max()
    angle_pad = max((angle_max - angle_min) * 0.25, 0.5)

    # Crear la figura con vista general + zoom del ángulo
    fig, (ax_main, ax_zoom) = plt.subplots(
        2,
        1,
        figsize=(12, 9),
        sharex=True,
        gridspec_kw={"height_ratios": [2, 1]},
    )

    # Vista general: tres señales sobrepuestas
    ax_main.plot(
        window_data["Time"],
        window_data["ECR_act"],
        label="ECR_act",
        alpha=0.8,
    )
    ax_main.plot(
        window_data["Time"],
        window_data["FCR_act"],
        label="FCR_act",
        alpha=0.8,
    )
    ax_main.plot(
        window_data["Time"],
        window_data["wrist_hand_r3_pos"],
        label="wrist_hand_r3_pos",
        alpha=0.8,
        color="green",
    )
    ax_main.set_title(f"Visualización de Señales - Ventana: {window_num}")
    ax_main.set_ylabel("ACTIVIDAD/ÁNGULOS")
    ax_main.grid(True, linestyle="--", alpha=0.6)
    ax_main.legend(loc="upper right")

    # Zoom específico del ángulo de muñeca
    ax_zoom.plot(
        window_data["Time"],
        window_data["wrist_hand_r3_pos"],
        color="green",
        label="wrist_hand_r3_pos (zoom)",
    )
    ax_zoom.set_title("Zoom del ángulo de muñeca")
    ax_zoom.set_xlabel("Tiempo (segundos)")
    ax_zoom.set_ylabel("Ángulo")
    ax_zoom.grid(True, linestyle="--", alpha=0.6)
    ax_zoom.set_ylim(angle_min - angle_pad, angle_max + angle_pad)
    ax_zoom.legend(loc="upper right")

    # Ajustar dinámicamente el límite del eje X
    t_start = start_idx / FS
    ax_zoom.set_xlim(t_start, t_start + WINDOW_DURATION)

    # Definir el nombre del archivo de la imagen y su ruta completa
    fig_name = f"{BASE_NAME}_ventana_{window_num:02d}.png"
    save_path = os.path.join(OUTPUT_DIR, fig_name)

    # Guardar la figura en alta calidad antes de mostrarla
    plt.savefig(save_path, dpi=300, bbox_inches="tight")

    print(
        f"Ventana {window_num} ({t_start}s a {t_start + WINDOW_DURATION}s) guardada en: {save_path}"
    )

    # Mostrar gráfico y pausar hasta que el usuario lo cierre
    plt.show()

print("Fin de las señales. Todas las figuras se han guardado correctamente.")
```
